```python
from sklearn.datasets import make_classification
```

```python
X,y= make_classification(n_samples=1000, n_features=2, n_redundant=0,n_clusters_per_class=1,weights=[0.90], random_state=123)
```

```python
df1= pd.DataFrame(X, columns=['f1','f2'])
df2= pd.DataFrame(y, columns=['target'])
```

```python
df= pd.concat([df1, df2], axis=1)
```

```python
plt.scatter(df['f1'], df['f2'], c= df['target'], alpha= 1)
```

```python
from imblearn.over_sampling import SMOTE
oversample= SMOTE()
```

```python
X,y= oversample.fit_resample(df[['f1','f2']],df['target'])
```

```python
df1= pd.DataFrame(X, columns=['f1','f2'])
df2= pd.DataFrame(y, columns=['target'])
oversample_df= pd.concat([df1,df2], axis=1)
```

```python
plt.scatter(oversample_df['f1'], oversample_df['f2'], c=oversample_df['target'])
```
